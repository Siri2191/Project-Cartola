```python
lista_times = ['São Paulo - SP', 'Palmeiras - SP', 'Fluminense - RJ',
       'Vasco da Gama - RJ', 'Grêmio - RS', 'Cruzeiro - MG', 'Avaí - SC',
       'Atlético - PR', 'Sport - PE', 'Chapecoense - SC',
       'Corinthians - SP', 'Santos - SP', 'Ponte Preta - SP',
       'Flamengo - RJ', 'Internacional - RS', 'Atlético - MG',
       'Figueirense - SC', 'Coritiba - PR', 'Goiás - GO',
       'Joinville - SC', 'América - MG', 'Santa Cruz - PE',
       'Vitória - BA', 'Botafogo - RJ', 'Bahia - BA', 'Atlético - GO',
       'Paraná - PR', 'Ceará - CE', 'CSA', 'Fortaleza', 'Bragantino - SP']

dicionario = {}

for item in lista_times:
  total_partidas_times = cartola_acumulado.query('VisitanteID=="{}" or CasaID=="{}"'.format(item,item))['Resultado'].count()
  vitoria_em_casa = cartola_acumulado.query('CasaID=="{}" & Resultado=="CASA"'.format(item))['Resultado'].count()
  vitoria_como_visitante = cartola_acumulado.query('VisitanteID=="{}" & Resultado=="VISITANTE"'.format(item))['Resultado'].count()
  total_vitorias = vitoria_como_visitante+vitoria_em_casa
  empates_como_visitante = cartola_acumulado.query('VisitanteID=="{}" & Resultado=="EMPATE"'.format(item))['Resultado'].count()
  empates_como_casa=cartola_acumulado.query('CasaID=="{}" & Resultado=="EMPATE"'.format(item))['Resultado'].count()
  total_empates = empates_como_casa + empates_como_visitante
  derrota_em_casa = cartola_acumulado.query('CasaID=="{}" & Resultado=="VISITANTE"'.format(item))['Resultado'].count()
  derrota_como_visitante = cartola_acumulado.query('VisitanteID=="{}" & Resultado=="CASA"'.format(item))['Resultado'].count()
  total_derrota = derrota_como_visitante + derrota_em_casa
  dicionario[item] = (total_partidas_times,
                      vitoria_em_casa,
                      vitoria_como_visitante,
                      total_vitorias,empates_como_visitante,
                      empates_como_casa,total_empates,derrota_em_casa,
                      derrota_como_visitante,total_derrota)
  score_por_times = pd.DataFrame(dicionario).transpose()

rename_columns={'index':'times',0:'total_partidas',1:'vitorias_em_casa',2:'vitoria_como_visitante',
                3:'total_vitorias',4:'empates_como_visitantes',5:'empates_em_casa',
                6:'total_empates',7:'derrotas_em_casa',8:'derrotas_como_visitante',9:'total_derrotas'}

score_por_times.rename(columns=rename_columns,inplace=True)
score_por_times
```

```python
score_por_times = score_por_times.reset_index()
score_por_times.rename(columns={'index':'times'},inplace=True)
score_por_times
```

```python

score_por_times.to_excel('score_por_time.xlsx', index='False')
score_por_times.to_csv('score_por_times.csv', index = 'False')
```
